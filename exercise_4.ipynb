# 1. Project Description:
Build a Financial/Investment Assistant that can answer queries using financial reports, market research, and economic data. Examples include stock analysis Q&A, SME financial advisory, and summarizing economic trends. The assistant should provide answers grounded in the retrieved context, with proper citations where necessary.

# 2. Would you use a framework or custom code? Why?
I would use a framework because the project requires:

- Fast prototyping
- Handling multiple documents and embeddings
- Easy integration with LLMs for summarization and Q&A
Using a framework reduces boilerplate and accelerates development while ensuring reliable RAG operations.

# 3. If using a framework, which one and why?
I would choose LlamaIndex because:

- The focus is on document question-answering and summarization, which aligns with financial data retrieval.
- It offers a simple API for indexing large financial reports and running queries.
- Optimized for RAG-specific tasks like semantic search and context-based Q&A.

# 4. What components would you keep custom (if any)?
I might keep prompt templates and response formatting custom:

- Designing prompts for financial analysis Q&A to enforce correct citation
- Formatting summaries or stock trend insights in a readable way
- Custom logic for alerting or highlighting important market trends

# 5. What are the main risks of your choice?

Framework dependency: future updates could change API behavior

- Limited flexibility: LlamaIndex is less flexible for multi-step workflows
- Data accuracy: relying on retrieved financial data means errors in reports could propagate
- Scalability: handling extremely large datasets may require tuning or hybrid solutions with custom code
